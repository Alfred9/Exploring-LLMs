```python
# Import required libraries and set hardware for model to use
import torch
from PIL import Image
from transformers import AutoProcessor, AutoModelForVision2Seq
from transformers.image_utils import load_image

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

print(DEVICE)
```

```python
# Download image from URL
import requests
url = "https://media.hswstatic.com/eyJidWNrZXQiOiJjb250ZW50Lmhzd3N0YXRpYy5jb20iLCJrZXkiOiJnaWZcL2JlY29taW5nLWRvY3Rvci5qcGciLCJlZGl0cyI6eyJyZXNpemUiOnsid2lkdGgiOjgyOH19fQ=="
image1 = Image.open(requests.get(url, stream=True).raw)
```

```python
# Load images
#image1 = load_image("https://www.google.com/url?sa=i&url=https%3A%2F%2Finfotradekenya.go.ke%2FRequirements%3Fpage%3D1%26letter%3Dk%26l%3Dsw&psig=AOvVaw1lrIbXJSV1u4ntldHmD66O&ust=1739283411702000&source=images&cd=vfe&opi=89978449&ved=0CBQQjRxqFwoTCND-t-GluYsDFQAAAAAdAAAAABAE")

print(image1)
```

```python
# Initialize processor and model
processor = AutoProcessor.from_pretrained("HuggingFaceTB/SmolVLM-Instruct")
model = AutoModelForVision2Seq.from_pretrained(
    "HuggingFaceTB/SmolVLM-Instruct",
    torch_dtype=torch.bfloat16
).to(DEVICE)
```

```python
# Create input messages
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image"},
            {"type": "text", "text": "what is going on in the picture?"}
        ]
    },
]
```

```python
# Prepare inputs
prompt = processor.apply_chat_template(messages, add_generation_prompt=True)
inputs = processor(text=prompt, images=[image1], return_tensors="pt")
inputs = inputs.to(DEVICE)

# print(inputs)
```

```python

# Generate outputs
generated_ids = model.generate(**inputs, max_new_tokens=500)
generated_texts = processor.batch_decode(
    generated_ids,
    skip_special_tokens=True,
)

print(generated_texts[0])
```
